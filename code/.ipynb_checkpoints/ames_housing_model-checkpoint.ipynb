```python
# Code: Take percentile data, remove % char, convert to float divide by 100.
def convert_ratio(value):
    
    get_num = value.replace('%', '')
    try:
        a = float(get_num)
        return a/100
    except ValueError:
        return np.nan
# This function get called to convert percentile to ratio
def str_float(perc):
    to_ratio = perc.apply(convert_ratio)
    return to_ratio
    
# take column name and update with lower and underscore if there is any space
def update_column(col):
    
    new_col = ''
    for c in col.strip().lower():
        if c.isalnum():
            new_col += c
        elif c == ' ':
            new_col += '_'
        elif c == '-':
            new_col += '_'
    return new_col

# Taking list of column and return dict of old and new column
def column_dict(col):

    new_col = {}
    for k in col:
        val = update_column(k)
        new_col[k] = val
    
    return new_col

# Taking dataset and return updated columns.
# the function should be called with dataset as arg to return updated columns.
def column_rename(dataset):
    
    old_cols = dataset.columns
    new_dict = column_dict(old_cols)
    dataset.rename(columns = new_dict, inplace=True)
    return dataset.columns

# update 'Yes' or 'No' to binary 1 | 0
def set_to_bin(col):

    # Take a column to replace Yes|No to 1|0
    replace_col = col.apply(lambda x: 1 if ('Yes' or 'yes') in x else 0 if ('No' or 'no') in x else np.nan).astype(float)
    return replace_col

# Return more than 25% of the column data that is null
def null_shape(dataset):
    data_len = dataset.shape[0]
    null_data = {}
    # Loop through the column and check if the column has null value more than 25% of row value
    for data in dataset:
        if dataset[data].isnull().sum() > 0.25 * data_len:
            null_data[data] = dataset[data].isnull().sum()
    return null_data

# unique categorical column 
def check_unique(dataset):
    
    unique_col = {}
    for row in dataset:
        unique_col[row] = dataset[row].unique().T
    return unique_col

# fill row with a given value
def fill_row(dataset, col, val):
    
    dataset[col] = dataset[col].fillna(val)
    return
def fill_category(dataset, val):
    
    for data in dataset:
        dataset[data] = dataset[data].fillna(val)
    return  


# Drop column 
def drop_column(dataset, col):
    
    return True

# high correlation filter (> 0.5) for dataset that has independent variables 
def high_corr(dataset, lim):
    
    highly_corr = []
    datacorr = dataset.corr()
    filtered = datacorr[((datacorr >= lim) | (datacorr <= -lim)) & (datacorr !=1.00)]
    filtered = filtered.unstack().sort_values(ascending=False).drop_duplicates()
    m = 0
    data_col =  dataset.columns
    data_len =  len(data_col)
    for data in dataset:
        m += 1 
        for i in range(data_len):
            n = i + m
            if n < data_len:
                col_name = data_col[n]
                cor = dataset[data].corr(dataset[col_name]) 
                if cor >= lim:
                    if data not in highly_corr:
                        highly_corr.append(data)
                    if col_name not in highly_corr:
                        highly_corr.append(col_name)
            
    return highly_corr, filtered

# print(train_numerical[train_numerical['mas_vnr_area'].isnull()].index.tolist())
```

```python
# modeling with pipeline
def test_modeling(X, y, models, test_size=0.3):

    # create empty dict to collect test results.
    results = [] 
    
    col = ['orange', 'blue', 'purple']
    i = 0
    # Go through each model and process the train/test split, find score, validate score and return the output
    for model in models:
        #print(models[model])
        output = {}
        # execute train/test split
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
        
        reg = models[model].fit(X_train, y_train)
        y_pred =  reg.predict(X_train)
        y_yhat = reg.predict(X_test)
        
        output['model'] = model
        output['train_score'] = reg.score(X_train, y_train)
        output['test_score'] = reg.score(X_test, y_test)
        # Training MSE
        output['train_mse']=mean_squared_error(y_train, y_pred)
        # Testing MSE
        output['test_mse'] = mean_squared_error(y_test, y_yhat)
        # Use the model in cross_val_score
        scores_train = cross_val_score(reg, X_train, y_train, cv=5)
        output['cross_val_train'] = scores_train
        scores_test = cross_val_score(reg, X_test, y_test, cv=5)
        output['cross_val_test'] = scores_test
        output['scores_mean_train'] = scores_train.mean()
        output['scores_std_train'] =  scores_train.std()
        output['scores_mean_test'] = scores_test.mean()
        output['scores_std_test'] =  scores_test.std()
        if 'GridSearchCV' in str(models[model]):
            print(f'Using GridSearch Best estimator for {model} Regression is {reg.best_estimator_}')
            print(f'Using GridSearch Best params for {model} Regression is {reg.best_params_}')
        
        results.append(output)
        plt.figure(figsize=(8, 6))
        plt.title("Predictions for training dataset")
        sns.regplot(x=y_pred, y = y_train, data=X_train, color=col[i], label=model)
        i += 1
    return results
        
        
```
